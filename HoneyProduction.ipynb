Is there any correlation between 2 numeric variables

```
sns.pairplot(mydata[['numcol' , 'yieldpercol' , 'totalprod' , 'stocks' , 'priceperlb' , 'prodvalue' , 'year']])
```

Calculating Correlation

```
cor=mydata[['numcol' , 
      'yieldpercol' , 
      'totalprod' , 
      'stocks' , 
      'priceperlb' , 
      'prodvalue' , 
      'year']].corr()
```

```
cor
```

Creating a heatmap

```
sns.heatmap(cor,annot=True,cmap='plasma',vmin=-1,vmax=1)
```

General production trend from 1998 to 2012

```
sns.pointplot(mydata['year'] , mydata['totalprod'])
```
